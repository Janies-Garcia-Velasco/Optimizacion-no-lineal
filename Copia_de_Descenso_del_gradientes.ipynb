```python
import numpy as np
import matplotlib.pyplot as plt

a=11
b=12
c=36
d=192
e=-2
f=-2

A=np.array([[a,b/2],[b/2,c]])
B=np.array([[d],[e]])
C=f

invA=np.linalg.inv(A)
xast=(1/2)*invA.dot(((-B)))

print(xast)

x1=xast[0]
x2=xast[1]
```

```python
def funcion(x1,x2):
  return a*(x1**2)+b*(x1*x2)+c*(x2**2)+d*(x1)+e*(x2)+f


x = np.arange(-30, 15)    # intervalo de la variable x
y = np.arange(-20, 20)    # intervalo de la variable y

X, Y = np.meshgrid(x, y)  # malla
Z = funcion(X,Y)

#plt.style.use('Solarize_Light2')
plt.contour(X,Y,Z, levels=20)
plt.plot(0,0,'ro')
plt.plot(x1,x2,"go")
plt.xlim([-10,10])
plt.xlim([-12,12])
plt.show()
```

```python
x0=np.random.randint(10, size=(2, 1))
N=0
alfa=0.01

lista = [x0]

while N < 20:
  f_x0=np.transpose(x0).dot(A).dot(x0)+np.transpose(x0).dot(B)+c
  grad_f_x0= 2.0 * A.dot(x0) + B
  x_1=x0-alfa*grad_f_x0
  N+=1

  lista.append(x_1)
  x0=x_1
  print(f_x0)

arreglo = np.array(lista).reshape((21,-1))
print(arreglo)
```
